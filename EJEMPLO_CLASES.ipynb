```python
from datetime import datetime

class Empleados:
    def __init__(self, nombre, apellidos, edad, posicion, anio, vacaciones, herramientas):
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad
        self.posicion = posicion
        self.anio = anio
        self.vacaciones = vacaciones
        self.herramientas = herramientas

    def descripcion(self):
        return f"{self.nombre} {self.apellidos} es {self.posicion}. Tiene {self.vacaciones} días de vacaciones restantes."

    def cambiando_posicion(self):
        for herramienta in self.herramientas:
            if herramienta.lower() == "python":
                return "¡Excelente! Tienes habilidades en Python."
            elif herramienta.lower() == "excel":
                return "Considera dejar de usar Excel y aprender Python."
        return "No tienes herramientas específicas mencionadas."

    def calcular_antiguedad(self):
        año_actual = datetime.now().year
        antiguedad = año_actual - self.anio
        return f"{self.nombre} lleva {antiguedad} años en la empresa."

    def solicitar_aumento_vacaciones(self, dias_solicitados):
        if dias_solicitados > 0:
            self.vacaciones += dias_solicitados
            return f"Solicitaste un aumento de {dias_solicitados} días. Ahora tienes {self.vacaciones} días de vacaciones."
        else:
            return "La solicitud debe ser un número positivo."

    def registrar_herramienta(self, nueva_herramienta):
        if nueva_herramienta not in self.herramientas:
            self.herramientas.append(nueva_herramienta)
            return f"Herramienta '{nueva_herramienta}' registrada con éxito."
        else:
            return "La herramienta ya está registrada."

    def cambiar_posicion(self, nueva_posicion):
        self.posicion = nueva_posicion
        return f"Posición cambiada a {self.posicion}."

    def tomar_vacaciones(self):
        while True:
            try:
                dias_solicitados = int(input("¿Cuántos días de vacaciones deseas tomar? "))
                if dias_solicitados <= 0:
                    print("Debes solicitar al menos un día de vacaciones.")
                    continue
                if dias_solicitados <= self.vacaciones:
                    self.vacaciones -= dias_solicitados
                    return f"Has tomado {dias_solicitados} días de vacaciones. Te quedan {self.vacaciones} días."
                else:
                    return "No tienes suficientes días de vacaciones."
            except ValueError:
                print("Por favor, ingresa un número válido.")

    def solicitar_aumento_salario(self):
        while True:
            try:
                aumento = float(input("¿Qué porcentaje de aumento deseas solicitar? "))
                if aumento < 0:
                    print("El porcentaje de aumento debe ser positivo.")
                    continue
                # Aquí asumimos que hay un salario base definido.
                # Solo para ilustrar, supongamos un salario base.
                salario_base = 30000
                nuevo_salario = salario_base * (1 + aumento / 100)
                return f"Tu nuevo salario sería: {nuevo_salario:.2f}"
            except ValueError:
                print("Por favor, ingresa un número válido.")

    def realizar_evaluacion(self):
        while True:
            calificacion = input("¿Cómo calificarías tu desempeño (excelente, bueno, regular, malo)? ").lower()
            if calificacion in ["excelente", "bueno", "regular", "malo"]:
                return f"Tu evaluación es: {calificacion}."
            else:
                print("Calificación no válida, intenta de nuevo.")

```
